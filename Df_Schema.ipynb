```python
# importing libraries

import twitter
import requests
import json
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
# bearer token

BEARER = 'XXX'
```

```python
# get user id

uname = 'blvckfoxmedia'

headers_res1 = {
    'Authorization': 'Bearer ' + BEARER,
}

response = requests.get('https://api.twitter.com/2/users/by/username/' + uname, headers=headers_res1)

response_json = json.loads(response.text)

uid = response_json['data']['id']
```

```python
# get phase dates

# Example Date: 14 May 2014

phase_1_date = '20 Jun 2022'
phase_2_date = '20 Oct 2022'
phase_3_date = '24 Mar 2023'

phase_1_date = datetime.strptime(phase_1_date, '%d %b %Y')
phase_2_date = datetime.strptime(phase_2_date, '%d %b %Y')
phase_3_date = datetime.strptime(phase_3_date, '%d %b %Y')
```

```python
# get phase tweets

params_1 = {
    'end_time': phase_1_date.isoformat()+'Z',
    'tweet.fields' : 'created_at,text,attachments',
    'expansions' : 'author_id,attachments.media_keys',
    'media.fields' : 'media_key,type,url,variants,preview_image_url',
    'max_results' : 100
    
}

params_2 = {
    'end_time': phase_2_date.isoformat()+'Z',
    'tweet.fields' : 'created_at,text,attachments',
    'expansions' : 'author_id,attachments.media_keys',
    'media.fields' : 'media_key,type,url,variants,preview_image_url',
    'max_results' : 100
    
}

params_3 = {
    'end_time': phase_3_date.isoformat()+'Z',
    'tweet.fields' : 'created_at,text,attachments',
    'expansions' : 'author_id,attachments.media_keys',
    'media.fields' : 'media_key,type,url,variants,preview_image_url',
    'max_results' : 100
    
}

headers_res2 = {
    'Authorization': 'Bearer ' + BEARER,
}

tweets_response_1 = requests.get('https://api.twitter.com/2/users/' + uid + '/tweets', params=params_1, headers=headers_res2)
tweets_response_2 = requests.get('https://api.twitter.com/2/users/' + uid + '/tweets', params=params_2, headers=headers_res2)
tweets_response_3 = requests.get('https://api.twitter.com/2/users/' + uid + '/tweets', params=params_3, headers=headers_res2)

tweets_json_1 = json.loads(tweets_response_1.text)
tweets_json_2 = json.loads(tweets_response_2.text)
tweets_json_3 = json.loads(tweets_response_3.text)
```

```python
# extracting info from tweets

mobs = []

for item in tweets_json_1['includes']['media']:
    
    row = []
    
    if 'photo' in str(item):
        row.append(item['media_key'])
        row.append('photo')
        row.append(item['url'])
    
    if 'video' in str(item):
        row.append(item['media_key'])
        for subitem in item['variants']:
            try:
                if subitem['bit_rate'] == 2176000:
                    row.append('video')
                    row.append(subitem['url'])
            except:
                pass

    mobs.append(row)

for item in tweets_json_2['includes']['media']:
    
    row = []
    
    if 'photo' in str(item):
        row.append(item['media_key'])
        row.append('photo')
        row.append(item['url'])
    
    if 'video' in str(item):
        row.append(item['media_key'])
        for subitem in item['variants']:
            try:
                if subitem['bit_rate'] == 2176000:
                    row.append('video')
                    row.append(subitem['url'])
            except:
                pass

    mobs.append(row)

for item in tweets_json_3['includes']['media']:
    
    row = []
    
    if 'photo' in str(item):
        row.append(item['media_key'])
        row.append('photo')
        row.append(item['url'])
    
    if 'video' in str(item):
        row.append(item['media_key'])
        for subitem in item['variants']:
            try:
                if subitem['bit_rate'] == 2176000:
                    row.append('video')
                    row.append(subitem['url'])
            except:
                pass

    mobs.append(row)

obs = []

for item in tweets_json_1['data']:
    row = [uid]
    row.append(item['created_at'])
    row.append(item['text'])
    
    try:
        if 'media_key' in str(item):
            if len(item['attachments']['media_keys']) == 1:
                for media_key in item['attachments']['media_keys']:
                    row.append(media_key)
            else:
                for i in range(len(item['attachments']['media_keys'])):
                    spl_row = [uid]
                    spl_row.append(item['created_at'])
                    spl_row.append(item['text'])
                    spl_row.append(item['attachments']['media_keys'][i])
                    obs.append(spl_row)
        else:
            row.append(0)
            
    except:
        pass
    
    obs.append(row)
    
for item in tweets_json_2['data']:
    row = [uid]
    row.append(item['created_at'])
    row.append(item['text'])
    
    try:
        if 'media_key' in str(item):
            if len(item['attachments']['media_keys']) == 1:
                for media_key in item['attachments']['media_keys']:
                    row.append(media_key)
            else:
                for i in range(len(item['attachments']['media_keys'])):
                    spl_row = [uid]
                    spl_row.append(item['created_at'])
                    spl_row.append(item['text'])
                    spl_row.append(item['attachments']['media_keys'][i])
                    obs.append(spl_row)
        else:
            row.append(0)
            
    except:
        pass
    
    obs.append(row)
    
for item in tweets_json_3['data']:
    row = [uid]
    row.append(item['created_at'])
    row.append(item['text'])
    
    try:
        if 'media_key' in str(item):
            if len(item['attachments']['media_keys']) == 1:
                for media_key in item['attachments']['media_keys']:
                    row.append(media_key)
            else:
                for i in range(len(item['attachments']['media_keys'])):
                    spl_row = [uid]
                    spl_row.append(item['created_at'])
                    spl_row.append(item['text'])
                    spl_row.append(item['attachments']['media_keys'][i])
                    obs.append(spl_row)
        else:
            row.append(0)
            
    except:
        pass
    
    obs.append(row)
```

```python
# create a dataFrame object

df = pd.DataFrame(columns = ['user_id' , 'created_at', 'text' , 'media_key'])
df = df.append(pd.DataFrame(obs,
               columns=[ 'user_id', 'created_at', 'text', 'media_key']),
               ignore_index = True)
df = df.dropna()
df.reset_index(inplace=True)

m_df = pd.DataFrame(columns = ['media_key' , 'content_type', 'url'])
m_df = m_df.append(pd.DataFrame(mobs,
               columns = ['media_key' , 'content_type', 'url']),
               ignore_index = True)

user_df = pd.merge(
    df, m_df, how='outer', on=['media_key']
)
user_df
```

```python
# categorizing phases

user_df['created_at'] = user_df['created_at'].map(lambda x: datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.000Z"))

conditions = [
    (user_df['created_at'] < phase_1_date),
    (user_df['created_at'] >= phase_1_date) & (user_df['created_at'] < phase_2_date),
    (user_df['created_at'] >= phase_2_date) & (user_df['created_at'] < phase_3_date),
    (user_df['created_at'] >= phase_3_date)
]

values = [1, 2, 3, 3]

user_df['phase'] = np.select(conditions, values)
```

```python
# dropping duplicates in cases of tweet overlaps and phase counts

user_df = user_df.drop_duplicates()

user_df['phase'].value_counts()
```

```python
import twitter
import requests
import json
from datetime import datetime

BEARER = 'XXX'

uname = input(str('Enter the username: '))
print('Example Date: 14 May 2014 \n')
sdate = input(str('Enter start date: '))
edate = input(str('Enter end date: \n'))

start_date = datetime.strptime(sdate, '%d %b %Y')
end_date = datetime.strptime(edate, '%d %b %Y')

headers_res1 = {
    'Authorization': 'Bearer ' + BEARER,
}

response = requests.get('https://api.twitter.com/2/users/by/username/' + uname, headers=headers_res1)

response_json = json.loads(response.text)

uid = response_json['data']['id']

params = {
    'start_time': start_date.isoformat()+'Z',
    'end_time': end_date.isoformat()+'Z',
    'tweet.fields' : 'text,created_at',
}

headers_res2 = {
    'Authorization': 'Bearer ' + BEARER,
}

tweets_response = requests.get('https://api.twitter.com/2/users/' + uid + '/tweets', params=params, headers=headers_res2)

tweets_json = json.loads(tweets_response.text)

for i in range(len(tweets_json['data'])):
    print('Created At: ' + tweets_json['data'][i]['created_at'])
    print('Text: ' + tweets_json['data'][i]['text'])
    print('-------------------------------------------------------------------------------------------------------------------')
```
